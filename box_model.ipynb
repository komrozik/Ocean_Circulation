```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
B = 100000
L = 100000
```

```python
def get_tau(y,L = 100000,tau_0 = 10**(-4)):
    return -tau_0*np.cos((np.pi*y)/(L))
```

```python
nx, ny = (25, 25)
x = np.linspace(0, B, nx)
y = np.linspace(0, L, ny)
xx, yy = np.meshgrid(x, y)
tau_x = get_tau(yy)
```

```python
plt.plot(xx, yy, marker='o', color='k', linestyle='none')
plt.show()
```

```python
plt.figure(dpi = 300)
h = plt.contourf(x, y, tau_x)
plt.quiver(x,y,tau_x,np.zeros(tau_x.shape))
plt.axis('scaled')
plt.colorbar()
plt.title("Wind Stress")
```

```python
def get_Stream(x,y
               ,L = 100000
               ,B = 100000
               ,tau_0 = 10**(-4)
               ,beta = 2*10**(-11)
               ):
    return (B-x)*(tau_0*np.pi)/(beta*L)*np.sin((np.pi*y)/(L))
```

```python
phi = get_Stream(xx,yy)

plt.figure(dpi = 300)
h = plt.contourf(x, y, phi)
#plt.quiver(x,y,tau_x,np.zeros(tau.shape))
plt.axis('scaled')
plt.colorbar()
plt.title("Streamfunction")
```
